### 1. Python Tuples
We introduced Tuples a little bit in the last class. Tuples are used to store multiple items in a single value.

#### Tuple items are ordered and unchangeable! --- different from a list

```python
studentname_tuple = "Mike", "Trevor", "Ben"
studentname_tuple = ("Mike", "Trevor", "Ben")

type(studentname_tuple)

print(studentname_tuple[1])

studentname_tuple[2] = "Emily"
```

However, you can change it to a list and then update it:

```python
studentname_list = list(studentname_tuple)
print(studentname_list)
studentname_list[2] = "Emily"
studentname_tuple = tuple(studentname_list)
print(studentname_tuple)
```

```python
# similar to a list, a tuple can be a combination of different types of elements
a = (1, 2, "xx")
type(a)
```

### 2. Python Sets

A set is a data type to store multiple items as a single variable.

#### set items are unordered and unchangeable!

```python
studentname_set = set(studentname_tuple)
print(studentname_set)
```

```python
# create a new set
colors_set = {"red", "blue", "yellow", "green"}
print(colors_set)
```

```python
# items in a set is not ordered, meaning index is not valid for a set
colors_set[1]
```

```python
# however, you can add or remove items from a set
# duplicate items will be included:
# if an item you want to add is already in the set, it will not be added
colors_set.add("white")
print(colors_set)
"""
colors_set.remove("blue")
print(colors_set)

colors_set.remove("purple")
print(colors_set)

colors_set.discard("purple")
print(colors_set)
"""
second_set = {"a", "b"}
colors_set.update(second_set)
print(colors_set)

second_list = ["c", "d"]
colors_set.update(second_list)
print(colors_set)

# use clear() to empty a set
# use del to delete a set
```

```python
# loop a set

for i in colors_set:
    print(i)

"""
# Does this work?

while len(colors_set) > 0:
    print(i)
"""
```

```python
# join sets
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)
print(set3)

set3 = set2.union(set1)
print(set3)

set1.update(set2)
print(set1)
```

```python
employee_rec = {
    "Name": "Min",
    "Dept": "FWE",
    "Sex": "Male",
    "Age": 37,
    "Title": "Assistant Professor"
}

x = employee_rec["Name"]
print(x)
x = employee_rec.get("Name")
print(x)

k = employee_rec.keys()
print(k)

employee_rec["Title"] = "Associate Professor"
print(employee_rec)

employee_rec.update({"Title": "Professor"})
print(employee_rec)

# add a key
employee_rec["Year"] = 2021
print(employee_rec)

employee_rec.update({"PI status": "Yes"})
print(employee_rec)
```

```python
# remove items

# pop removes the item with the specified key
employee_rec.pop("Sex")
print(employee_rec)

# or you can use del
del employee_rec["Dept"]
print(employee_rec)

# popitem removes the last inserted item
employee_rec.popitem()
print(employee_rec)

# similar to Set, del dictname and clear() delete the whole dictionary and empty the dictionary
# employee_rec.clear()
# del employee_rec
```

```python
# loop dictionary
employee_rec = {
    "Name": "Min",
    "Dept": "FWE",
    "Sex": "Male",
    "Age": 37,
    "Title": "Assistant Professor"
}

for x in employee_rec:
    print(x)

print("\nkeys:")
for x in employee_rec.keys():
    print(x)

print("\nvalues:")
for x in employee_rec.values():
    print(x)

print("\nitems:")
for x in employee_rec.items():
    print(x)
```

```python
# merge two dictionaries
dic_a = {'A': 'Apple', 'B': 'Ball'}
dic_b = {'C': 'Cat', 'D': 'Dog'}
dic_a.update(dic_b)
print(dic_a)

# easier way in python 3.9 and above
dic_a = {'A': 'Apple', 'B': 'Ball'}
dic_b = {'C': 'Cat', 'D': 'Dog'}
dic_c = dic_a | dic_b
print(dic_c)

# Concatenating more than 2 dictionaries
dic_d = dic_a | dic_b | dic_c
print(dic_d)
```
