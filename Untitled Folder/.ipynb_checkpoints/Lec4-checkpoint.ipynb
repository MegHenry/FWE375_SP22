```python
# paper scissors rock
import random
random.seed()
a = random.randint(1,3)
b = random.randint(1,3)
if a > b:
    print("a win!")
else:
    print("b win!")
print("a=", a, "b=", b)
```

```python
import random
random.seed()
a = random.randint(1,3)
b = random.randint(1,3)
if a > b:
    print("a win!")
elif a == b:
    print("nobody wins.")
else:
    print("b win!")
print("a=", a, "b=", b)
```

From Python 3.10, "match...case..." is supported

```python
# a match ... case ... example

import random
random.seed()
a = random.randint(1,100)

a = 1
match a:
    case 1:
        print("John!")
    case 2:
        print("Brendon!")
    case 3:
        print("Megan!")
    case 4:
        print("Nick!")
    case 5 or 6:
        print("Morgan!")
    case _:
        print("I am tired!")
```

## Loop is very useful because we are always asking computer to repeat something.
- for loop
- while loop

```python
for i in range(1,10):
    print(i)
```

```python
i = 1
while i in range(1,10):
    i = i + 1
    print(i)
    
```

```python
for i in [10,23,45,87]:
    i += 100
    print(i)
```

```python
tot = 0
for i in range(1,101):
    tot = tot + i
print(tot)
```

```python
# create a Fibonacci sequence
N = 100
L = [0, 1]
for i in range(1, N+1):
    L = L + [L[-2] + L[-1]] 
print(L)
```

```python
import random
random.seed()

L = random.sample(range(1,101), 10)
print(L)


```

```python
import random
random.seed()
a_win = 0
b_win = 0
no_win = 0
for i in range(1,100):
    a = random.randint(1,3)
    b = random.randint(1,3)
    if a > b:
        a_win = a_win + 1
    elif a == b:
        no_win += 1
    else:
        b_win = b_win + 1

print("a_win=", a_win, "b_win=", b_win, "no_win=", no_win)
```

```python
# calculate pi
import random

random.seed()
in_circle = 0
out_circle = 0
for i in range(1,100000):
    x = random.randrange(-1000,1000)/1000
    y = random.randrange(-1000,1000)/1000
    if x**2 + y**2 <= 1:
        in_circle = in_circle + 1
    else:
        out_circle = out_circle + 1

print(in_circle, "points are in the circle,", out_circle, "points are out of the circle.", out_circle)
print("estimated pi equals ", in_circle/(in_circle+out_circle)*4)
```
