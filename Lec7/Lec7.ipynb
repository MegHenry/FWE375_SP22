```python
# to use numpy, we need to import the numpy package
import numpy as np

# ways to create a numpy array
A = np.array([1, 2, 3.4, 5])
print(A)
print(type(A))

A = np.arange(0, 5, 1, dtype=int)
print(A)

A = np.arange(0, 5, 0.5, dtype=int)
print(A)

A = np.linspace(0, 10, 5, endpoint=False)
print(A)

A = np.zeros(10)
print(A)

A = np.ones(10) * 10
print(A)

B = np.empty(10)
B[:] = np.nan
print(B)
```

```python
# create a 2D array
A = np.array([[1, 2, 3, 4],[2, 4, 5, 6]])
print(A)
print(type(A))
print(A.shape)
print(np.nanmean(A))
print(np.sum(A))
print(np.median(A))
print(np.std(A))
print(np.mean(A))

# what if there is a nan value in A?
```

```python
# index for a 2D array
A = np.array([[1, 2, 3, 4],[2, 4, 5, 6]])
A[0,1]
A[1,]
B = A[:,2]
B.shape
```

```python
# math operations
A = np.linspace(0, 10, 11)
print(A)
B = A + 3
print(B)
C = A * 1.3 + 20
print(C)
```

```python
# read a 2D dataset
fname = "globalelev.txt"
elev = np.loadtxt(fname)

import seaborn as sns
ax = sns.heatmap(elev)
```

```python
# find the highest place in the world
x, y = np.where(elev == np.amax(elev))
Lat = np.linspace(90, -90, 720)
Lon = np.linspace(-180, 180, 1440)
print(Lat[x], Lon[y])
```
